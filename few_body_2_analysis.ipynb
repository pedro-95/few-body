# Deuteron Problem

```python
def AnaliticalSolution(landa, A):
    return (np.pi/2 + np.arctan(1/np.sqrt(landa - 1)))/(A*np.sqrt(landa-1))
```

```python
N = 2                                  # Number of Nuclei Particles
B_d = 2.226                            # Binding energy of the Deuteron in MeV
c = 2.998e+10                          # Speed of Light in cm/s
M = 938.272081                         # Mass of Proton in MeV/cÂ²
hbar = 6.5821196e-22                   # h bar constant in MeV*s
R_d = hbar / (2*(M/c**2)*B_d)**(1/2)   # Unit of length in cm
A =  R_d/(hbar/np.sqrt(M*B_d/(c)**2))
```

```python
ini = 20-1
end = 35

var_holder = {}
 
for i in range(ini,end):
    var_holder['landa_' + str(i+1)] = np.loadtxt("samples/landa_" + str(i+1))
 
locals().update(var_holder)
```

```python
landa_deut_mean  = np.zeros(end-ini)
landa_deut_error = np.zeros(end-ini)
j=0

for i in range(ini,end):
    landa_deut_mean[j], landa_deut_error[j] = SimpleBlockBootstrap(var_holder['landa_' + str(i+1)])
    j+=1
    print ('landa_' + str(i+1), end="\r")
```

```python
landa = np.linspace(3, 20., 10000)
b = AnaliticalSolution(landa,A)

b_deut = np.arange(0.6,2.2,0.1)


rc('font',**{'family':'sans-serif','sans-serif':['Helvetica']})
## for Palatino and other serif fonts use:
#rc('font',**{'family':'serif','serif':['Palatino']})
rc('text', usetex=True)
plt.rcParams.update({'font.size': 22})

plt.figure(figsize=(10,10))
plt.plot(b,landa, label = "Analytical solution of the Deuteron (N = 2)")
plt.errorbar(b_deut,landa_deut_mean, yerr=landa_deut_error,fmt="ro", ecolor = "k", label = "Montecarlo integration")
plt.xlabel(r"$b/R$")
plt.ylabel(r"$V_{0} / B$")
plt.legend()
plt.savefig("plots/Deuteron.pdf", bbox_inches="tight")
plt.show()
```

```python
def PointEstimator(R_m):
    L, = np.shape(R_m)
    N = int(L/3) 
    
    suma = 0.
    norma = 0.
    
    for i in range (0 , N):
        for j in range (i+1, N):
            suma += norm(R_m[3*i:3*(i+1)]-R_m[3*j:3*(j+1)])
            norma += 1.
    return suma/norma

def WaveFunctionEstimator(R):
    size, L = np.shape(R)
    wave = np.zeros(size)
    
    for i in range (0,size):
        wave[i] = PointEstimator(R[i])
        
    return wave
```

```python
from mpl_toolkits.axes_grid1.inset_locator import inset_axes


def test_hist_2d(x_data, y_data, x_labels, y_labels, Name_of_the_particle, save_fig = False):
    nrows, ncols = 1, 3
    fig, axes = plt.subplots(nrows, ncols, sharex=True, sharey=True,constrained_layout=True, figsize = (30,10))
    
    fig.suptitle(Name_of_the_particle, fontsize=50)
    
    ##produce the actual data and compute the histograms
    mappables=[]
    for (i,), ax in np.ndenumerate(axes):
        H, xedges, yedges,image = plt.hist2d(x_data[i], y_data[i], bins=50, density=True , range=[[-20, 20],[-20, 20]])
        ax.set_xlabel(x_labels[i], fontsize =40)
        ax.set_ylabel(y_labels[i], fontsize =40)
        ax.set_aspect("equal")
        mappables.append(H)

    ##the min and max values of all histograms
    vmin = np.min(mappables)
    vmax = np.max(mappables)
    #vmax = 0.026
    
    ##second loop for visualisation
    for ax, H in zip(axes.ravel(), mappables):
        im = ax.imshow(H,vmin=vmin, vmax=vmax, extent=[-20,20,-20,20])#, cmap = "seismic"

    ##colorbar using solution from linked question
    axins = inset_axes(ax,
                    width="5%",  # width = 50% of parent_bbox width
                    height="100%",  # height : 5%
                    loc='upper right',
                    bbox_to_anchor=(0.2, 0.01, 1., 1),
                    bbox_transform=ax.transAxes,)

    
    fig.colorbar(im ,ax=axes.ravel(), cax = axins)
    
    if (save_fig == True):
        plt.savefig("plots/"+Name_of_the_particle+".pdf", bbox_inches="tight")
    plt.show()
```

```python
def DeuteronPointEstimator(R_m):
    L, = np.shape(R_m)
    N = int(L/3) 
    suma_1 = 0.
    suma_2 = 0.
    suma_3 = 0.
    norma = 0.
    for i in range (0 , N):
        for j in range (i+1, N):
            suma_1 += R_m[3*i]-R_m[3*j]
            suma_2 += R_m[3*i+1]-R_m[3*j+1]
            suma_3 += R_m[3*i+2]-R_m[3*j+2]
            norma+=1.
    return N*suma_1/norma, N*suma_2/norma, N*suma_3/norma

def DeuteronWaveFunctionEstimator(R):
    size, L = np.shape(R)
    wave = np.zeros((size,3))
    
    for i in range (0,size):
        wave[i] = DeuteronPointEstimator(R[i])
    return wave
```
